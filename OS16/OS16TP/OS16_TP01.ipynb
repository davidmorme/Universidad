Création des données linéairement séparables

```python
nbi=100
x1=randn(nbi,2)+[1,2]
x2=randn(nbi,2)-[1,2]

xapp=np.concatenate((x1, x2), axis=0)

lab_app=np.zeros(2*nbi)
lab_app[0:nbi]=1
lab_app[nbi:2*nbi]=-1
```

```python
plt.scatter(xapp[0:nbi,0],xapp[0:nbi,1])
plt.scatter(xapp[nbi:2*nbi,0],xapp[nbi:2*nbi,1],c='r')
plt.xlabel('x')  
plt.ylabel('y')  
  
# displaying the title 
plt.title("samples") 
plt.show()
```

Création des données de test

```python
nbt=1000
x1=randn(nbt,2)+[1,2]
x2=randn(nbt,2)-[1,2]

xt=np.concatenate((x1, x2), axis=0)

lab_test=np.zeros(2*nbt)
lab_test[0:nbt]=1
lab_test[nbt:2*nbt]=-1
```

```python
plt.scatter(xt[lab_test>0,0],xt[lab_test>0,1])
plt.scatter(xt[lab_test<0,0],xt[lab_test<0,1],c='r')
plt.xlabel('x')  
plt.ylabel('y')  
  
# displaying the title 
plt.title("test samples") 
plt.show()
```

Affichage des deux classes

```python
### Entrenamiento del modelo ###
classifier = svm.SVC(kernel='linear')
#classifier = svm.SVC(kernel='rbf',gamma=0.5)
classifier.fit(xapp, lab_app)
```

```python
#Predicción para un punto específico en el espacio
classifier.decision_function([[-2,-2]])
```

```python
y_app_pred = classifier.decision_function(xapp)
lab_app_pred = np.sign(y_app_pred)
y_test_pred = classifier.decision_function(xt)
lab_test_pred = np.sign(y_test_pred)

eval_app=lab_app_pred!=lab_app
eval_test=lab_test_pred!=lab_test
```

```python
print('Datos de aprendizaje mal clasificados:', sum(eval_app)/nbi/2*100,'%')
print('Datos de Test mal clasificados:',sum(eval_test)/nbt/2*100,'%')
```

```python
from sklearn import metrics
fpr, tpr, thr = metrics.roc_curve(y_true=lab_test, y_score=y_test_pred)

# calculer l'aire sous la courbe ROC
auc = metrics.auc(fpr, tpr)

# créer une figure
fig = plt.figure(figsize=(6, 6))

# afficher la courbe ROC
plt.plot(fpr, tpr, '-', lw=2, label='Lineaire, AUC=%.2f' % auc)

# donner un titre aux axes et au graphique
plt.xlabel('Taux de faux Positif', fontsize=16)
plt.ylabel('Taux de vrai Positif', fontsize=16)
plt.title('SVM Courbe COR, AUC = %.3f'%auc, fontsize=16)

# afficher l'image
plt.show()
```

Parcours d'une grille 2D pour affichage de la valeur de la fonction SVM

```python
origin = 'lower'
nbs=100
delta=10/(nbs-1)
x = y = np.arange(-5.0, 5.0001, delta)
X, Y = np.meshgrid(x, y)
Xr=np.reshape(X,(nbs*nbs,1))
Yr=np.reshape(Y,(nbs*nbs,1))
Xmap=np.concatenate((Xr, Yr), axis=1)
Ymapr = classifier.decision_function(Xmap)
Ymap=np.reshape(Ymapr,(nbs,nbs))
Ymap.shape

fig1, ax1 = plt.subplots(constrained_layout=True)
CS = ax1.contourf(X, Y, Ymap, 10, cmap=plt.cm.bone, origin=origin)
ax1.set_title('SVM output heatmap')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

#CS2 = ax1.contour(CS, levels=CS.levels[::1], colors='r', origin=origin)
CS2 = ax1.contour(CS, levels=(-1,0,1), colors='r', origin=origin)
# Make a colorbar for the ContourSet returned by the contourf call.
cbar = fig1.colorbar(CS)
cbar.ax.set_ylabel('Output value')
# Add the contour line levels to the colorbar
cbar.add_lines(CS2)

ax1.scatter(xapp[lab_app<0,0],xapp[lab_app<0,1])
ax1.scatter(xapp[lab_app>0,0],xapp[lab_app>0,1],c='r')

plt.show()
```
