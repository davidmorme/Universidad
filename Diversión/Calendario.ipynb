###Código completo###

```python
import numpy as np

def calculadoraDiaSemana(mes, año):
  diasMeses = {1 : 31, 2 : 28, 3: 31, 4: 30, 5: 31, 6:30, 7:31, 8:31, 9:30, 10:31, 11: 30, 12:31}
  encabezado=["D","L","M","M","J","V","S"]
  calendario=np.zeros((6,7))
  semana=0
  if mes <= 2:
    diaNumero = (año + 31 * (mes - 1) + 1*(año - 1) // 4 - 3 *((año + 99) // 100) // 4) %7
  else:
    diaNumero = (año + 31 * (mes - 1) + 1 - (4 * mes + 23) // 10 + año // 4 - (3 * (año // 100 + 1))// 4-1)%7
  for dia in range( 1 , diasMeses[mes]+1):
    calendario[semana,diaNumero]=dia
    if(diaNumero==6):
      semana+=1
      diaNumero=0
    else:
      diaNumero+=1
  print(encabezado,'\n',calendario)

while True:
    mes = input('Ingrese el mes ')
    try:
        mes = int(mes)
    except:
        print('Por favor use un número válido (de 1 a 12)')
        continue
    if mes < 1 or mes > 12:
        print('Por favor use un mes válido (de 1 a 12)')
        continue
    break
while True:
    año = input('Ingrese el año ')
    try:
        año = int(año)
    except:
        print('Por favor use un número válido para el año')
        continue
    if año < -4.5*10**9:
        print('Por favor use un número válido, excediste los 4,500 millones de años que tiene la tierra')
        continue
    break

calculadoraDiaSemana(mes, año)
```
