```python
import numpy as np

def calculadoraDiaSemana(mes, ao):
  diasMeses = {1 : 31, 2 : 28, 3: 31, 4: 30, 5: 31, 6:30, 7:31, 8:31, 9:30, 10:31, 11: 30, 12:31}
  encabezado=["D","L","M","M","J","V","S"]
  calendario=np.zeros((6,7))
  semana=0
  if mes <= 2:
    diaNumero = (ao + 31 * (mes - 1) + 1*(ao - 1) // 4 - 3 *((ao + 99) // 100) // 4) %7
  else:
    diaNumero = (ao + 31 * (mes - 1) + 1 - (4 * mes + 23) // 10 + ao // 4 - (3 * (ao // 100 + 1))// 4-1)%7
  for dia in range( 1 , diasMeses[mes]+1):
    calendario[semana,diaNumero]=dia
    if(diaNumero==6):
      semana+=1
      diaNumero=0
    else:
      diaNumero+=1
  print(encabezado,'\n',calendario)

while True:
    mes = input('Ingrese el mes ')
    try:
        mes = int(mes)
    except:
        print('Por favor use un nmero vlido (de 1 a 12)')
        continue
    if mes < 1 or mes > 12:
        print('Por favor use un mes vlido (de 1 a 12)')
        continue
    break
while True:
    ao = input('Ingrese el ao ')
    try:
        ao = int(ao)
    except:
        print('Por favor use un nmero vlido para el ao')
        continue
    if ao < -4.5*10**9:
        print('Por favor use un nmero vlido, excediste los 4,500 millones de aos que tiene la tierra')
        continue
    break

calculadoraDiaSemana(mes, ao)
```
