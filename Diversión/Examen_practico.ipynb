After, I verified the information of the dataset such as the amount of null values in each column, realizing that there are not missing values in this dataset.

Looking at the columns:

The following analysis will be done with a new dataset of **833** rows, gotten from subsetting the original dataset by excluding the pet types rabbit and hamster.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
pets=pd.read_csv('Archivos/pet_sales.csv')
pets.head()
```

```python
pets.sales=pets.sales.str.replace('$','',regex=True).replace(',','',regex=True).astype('float')
print(pets.info(), '\n')
print(pets.describe(), '\n')
```

```python
print('Cantidad de ids únicos: ',len(pets.product_id.unique()))
print('Cantidad de vendedores ids únicos: ',len(pets.vendor_id.unique()))
print('Tipos de categorías: ',pets.product_category.unique())
print('Únicos en tamaño: ',pets.pet_size.unique())
print('Únicos en tipo: ',pets.pet_type.unique())
print('Únicos en re buy: ',pets.re_buy.unique())
```

```python
SubPets=pets[pets.pet_type.isin(['fish', 'cat' , 'dog', 'bird'])]
print('Únicos en tipo: ',SubPets.pet_type.unique())
display(SubPets.head())
```

```python
print(SubPets.info(), '\n')
print(SubPets.describe(), '\n')
```

```python
def func(pct, allvals):
    absolute = int(np.round(pct/100.*np.sum(allvals)))
    return "{:.1f}%\n({:d} products)".format(pct, absolute)

cant=SubPets['re_buy'].value_counts()
fig,ax=plt.subplots(figsize=(4.5,4.5))
ax.pie(cant.values,labels=['No','Yes'], autopct=lambda pct: func(pct, cant), explode=(0.02, 0.02), shadow=True,startangle = 90)
ax.set_title('Quantity of products that are being \npurchased more than once',fontsize=11)
fig.savefig('Archivos/Rebuy_pie.png')
fig.show
```

```python
fig,ax=plt.subplots(figsize=(7,7))
sns.boxplot(x='product_category',y='sales',hue='re_buy',data=SubPets)
plt.xticks(rotation = 45)
ax.set_title('Sales of products by category and if are \npurchased more than once',fontsize=11)
ax.set_ylabel('Sales')
ax.set_xlabel('Category')
fig.savefig('Archivos/Sales_Category_rebuy.png')
fig.show
```

```python
fig,ax=plt.subplots()
sns.violinplot(x='re_buy',y='rating',data=SubPets)
plt.xticks([0,1],['No','Yes'])
ax.set_title('Sales of products by category and if are \npurchased more than once',fontsize=11)
ax.set_ylabel('Rating')
ax.set_xlabel('Rebuy products')
#fig.savefig('Archivos/Sales_Category_rebuy.png')
fig.show
```

```python
SubPets[['re_buy','rating']].groupby('re_buy').describe()
```

```python
fig,ax=plt.subplots()
sns.histplot(x='rating',data=SubPets[SubPets['re_buy']==0],kde=True,stat="density",bins=10, label='Non rebuy products')
sns.histplot(x='rating',data=SubPets[SubPets['re_buy']==1],kde=True,stat="density",bins=10, color='orange', label='Rebuy products')
ax.set_title('Rating of products if are \npurchased more than once',fontsize=11)
ax.set_xlabel('Rating')
ax.legend()
fig.savefig('Archivos/Rating_rebuy.png')
fig.show
```

```python
#What products are more likely to be purchased again for different types of pets?
SubPets['Quantity']=SubPets['sales']/SubPets['price']
SubPets[SubPets['re_buy']==1].sort_values('Quantity',ascending=False).head()
```

```python
birds=SubPets[(SubPets['re_buy']==1) & (SubPets['pet_type']=='bird')].sort_values('Quantity',ascending=False).reset_index(drop=True)
cats=SubPets[(SubPets['re_buy']==1) & (SubPets['pet_type']=='cat')].sort_values('Quantity',ascending=False).reset_index(drop=True)
dogs=SubPets[(SubPets['re_buy']==1) & (SubPets['pet_type']=='dog')].sort_values('Quantity',ascending=False).reset_index(drop=True)
fishs=SubPets[(SubPets['re_buy']==1) & (SubPets['pet_type']=='fish')].sort_values('Quantity',ascending=False).reset_index(drop=True)
```

```python
birds.head()
```

```python
maxs=SubPets[SubPets['re_buy']==1].groupby('pet_type')[['product_id','Quantity']].max()

fig,ax=plt.subplots(figsize=(4.5,4.5))

bars=sns.barplot( y=maxs.Quantity,x=maxs.product_id, order=maxs.product_id)
for i in range(len(maxs)):
    ax.text(i-0.15,maxs.Quantity[i]+5*10e2, maxs.index[i])
ax.set_title('Top products in units sold by pet type',fontsize=11)
#ax.set_yticks([1e5,2e5,3e5,4e5],['100m','200m','300m','400m','50M'])

ax.set_ylabel('Total units sold')
ax.set_xlabel("Product's id")
fig.savefig('Archivos/Top_products.png')
fig.show
```

```python
fig,ax=plt.subplots(figsize=(4.5,4.5))

counts=SubPets.value_counts('pet_type')

sns.barplot( y=counts.values,x=counts.index, order=maxs.index)
ax.set_title('Amount of products per pet type',fontsize=11)
ax.set_ylabel('Amount of products')
ax.set_xlabel('Pet type')
#fig.savefig('Archivos/TotalSales_Rebuy.png')
fig.show
```
